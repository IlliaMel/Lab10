```python
df = pd.read_csv('train.csv')
```

# IІ. Data Visualization.

2. (10б) Створіть pie chart, який би показував співвідношення виживших до загиблих.

```python
plt.figure(figsize=(5, 5))
plt.pie(df['Survived'].value_counts(), labels=['Not Survived', 'Survived'], autopct='%1.2f%%', startangle=90)
plt.title('Ratio of survivors to dead')
plt.show()
```

So, according to the obtained data, it can be understood that most of the passengers did not survive, namely 61.62%. However, the other part, namely 38.38%, managed to survive

3. (20б) Створіть bar chart, який би показував співвідношення загиблих до виживших для кожної статі. Для groupby використовуйте атрибут as_index=False.


image.png

```python
survived = df.groupby(['Sex', 'Survived'], as_index=False).size().pivot(index='Sex', columns='Survived', values='size')

survived.plot(kind='bar', stacked=True, color=['blue', 'orange'])

plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survived vs Dead by Sex')
plt.legend(title='Survived', labels=['Dead', 'Survived'], loc='upper right')
plt.xticks(rotation=0)
plt.show()
```

Analyzing these data, you can see that there were fewer women, but more of them survived, perhaps this is due to the fact that women and children were initially evacuated

4. (20б) Відобразіть кількість пропущених значень в датасеті по кожній із змінних.

image.png

```python
values = df.isnull().sum().sort_values(ascending=False)

plt.figure(figsize=(6, 4))
values.plot(kind='bar', color='blue')
plt.title('The amount of nullable values per column')
plt.xticks(rotation=45, ha='right')
plt.show()
```

From the analysis of these data, it can be understood that the place of each of the passengers is mostly unknown

5. (10б) Побудуйте наступний графік, дайте йому назву та опишіть, що власне на них відображається.

image.png

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df, palette='Set1', split=True)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.legend(title='Survived', loc='best', labels=['No', 'Yes'])
plt.show()
```

This diagram shows how the class of the ticket and the age of the passenger affected whether it survived or not

6. (13б) Побудуйте графік розподілу частот по віку пасажирів.



```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.axvline(df['Age'].median(), color='red', linestyle='dashed', linewidth=1, label='Median age')
plt.axvline(df['Age'].mean(), color='green', linestyle='dashed', linewidth=1, label='Mean age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Passenger Age Distribution')
plt.legend()
plt.show()
```

This visualization shows us that the average age and the median are very close, it can also be understood that most of the passengers were middle-aged

7. (25б) Поставте бізнес-питання до даних. Дайте на нього відповідь за допомогою візуалізацій. Прокоментуйте отриману відповідь.

The dependence of the chance of survival on age and gender

```python
age_bins = pd.cut(df['Age'], bins=6)  # Розділення віку на 10 бінів
age_bins.name = 'Age Group'
df['Age Group'] = age_bins

plt.figure(figsize=(12, 6))
sns.heatmap(df.groupby(['Age Group', 'Sex'])['Survived'].mean().unstack(), annot=True, fmt=".2f", cmap='coolwarm')
plt.xlabel('Sex')
plt.ylabel('Age Group')
plt.title('Survival Chance by Age Group and Sex')
plt.show()
```

With the help of these diagrams, you can understand the probability of survival of a person depending on his gender and age
